```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_in_use = Path("Resources/schools_complete.csv")
student_data_in_use = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_in_use)
student_data = pd.read_csv(student_data_in_use)

# Combine the data into a single dataset.
final_student_df = pd.merge(student_data, school_data, on=["school_name", "school_name"])
final_student_df.head()
```

```python
# Calculate the total number of unique schools
number_of_schools = len(final_student_df["school_name"].unique())
print("Number of Schools:",number_of_schools)
```

```python
# Calculate the total number of students
number_of_students = final_student_df["school_name"].value_counts().sum()
print(f"Number of Students: {number_of_students:,}")
```

```python
# Calculate the total budget
total_budget = final_student_df["budget"].unique().sum()
print(f"Total Budget: ${total_budget:,.2f}")
```

```python
# Calculate the average (mean) math score
math_average_score = final_student_df["math_score"].mean()
math_average_score = math_average_score
print(f"Average Math Score: {math_average_score:.2f}")
```

```python
# Calculate the average (mean) reading score
reading_average_score = final_student_df["reading_score"].mean()
print(f"Average Reading Score: {reading_average_score:.2f}")
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
pass_math_students = final_student_df[(final_student_df["math_score"] >= 70)].count()["student_name"]
math_pass_rate = (pass_math_students / number_of_students) * 100
print(f"Math Passing Rate: {math_pass_rate:.2f}%")
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
pass_reading_students = final_student_df[(final_student_df["reading_score"] >= 70)].count()["student_name"]
reading_pass_rate = (pass_reading_students / number_of_students) * 100
print(f"Reading Passing Rate: {reading_pass_rate:.2f}%")
```

```python
# Use the following to calculate the percentage of students that passed math and reading
pass_math_reading = final_student_df[(final_student_df["math_score"] >= 70) & (final_student_df["reading_score"] >= 70)].count()["student_name"]
math_reading_pass_rate = (pass_math_reading / number_of_students) * 100
print(f"Overall Passing Rate: {math_reading_pass_rate:.2f}%")
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Number of Schools": [number_of_schools],
    "Number of Students": [f"{number_of_students:,}"],
    "Total Budget": [f"${total_budget:,.2f}"],
    "Average Math Score": [f"{math_average_score:.2f}"],
    "Average Reading Score": [f"{reading_average_score:.2f}"],
    "Math Passing Rate": [f"{math_pass_rate:.2f}%"],
    "Reading Passing Rate": [f"{reading_pass_rate:.2f}%"],
    "Overall Passing Rate": [f"{math_reading_pass_rate:.2f}%"]
})
district_summary
```

```python
school_names = sorted(final_student_df["school_name"].unique())
school_names
```

```python
school_types = final_student_df.groupby("school_name")["type"].first()
school_types
```

```python
students_per_school = final_student_df["school_name"].value_counts()
students_per_school
```

```python
budget_per_school = final_student_df.groupby("school_name")["budget"].first().sort_values(ascending=False)
budget_per_school
```

```python
per_capita_budget = (budget_per_school / students_per_school).sort_values(ascending=False)
per_capita_budget
```

```python
school_average_math = final_student_df.groupby("school_name")["math_score"].mean().sort_values(ascending=False)
school_average_math
```

```python
school_average_reading = final_student_df.groupby("school_name")["reading_score"].mean().sort_values(ascending=False)
school_average_reading
```

```python
school_pass_math = ((final_student_df[final_student_df["math_score"] >= 70].groupby("school_name").count()["student_name"] / students_per_school) * 100).sort_values(ascending=False) 
school_pass_math
```

```python
school_pass_reading = ((final_student_df[final_student_df["reading_score"] >= 70].groupby("school_name").count()["student_name"] / students_per_school) * 100).sort_values(ascending=False) 
school_pass_reading
```

```python
both_pass_school = final_student_df[(final_student_df["math_score"] >= 70) & (final_student_df["reading_score"] >= 70)]
school_pass = both_pass_school.groupby("school_name").count()['student_name']
school_pass_rate = ((school_pass / students_per_school) * 100).sort_values(ascending=False)
school_pass_rate
```

```python
per_school_summary = pd.DataFrame({
    "School Name": school_names,
    "School Type": school_types,
    "Student Count": students_per_school,
    "Budget ($)": budget_per_school,
    "Per Capita Budget ($)": per_capita_budget,
    "Math Average Score": school_average_math,
    "Reading Average Score": school_average_reading,
    "Math Passing Rate (%)": school_pass_math,
    "Reading Passing Rate (%)": school_pass_reading,
    "Overall Passing Rate (%)": school_pass_rate})

per_school_summary.set_index("School Name", inplace=True)
per_school_summary
```

```python
top_schools = per_school_summary.sort_values(by = "Overall Passing Rate (%)", ascending = False)
top_schools.head()
```

```python
bottom_schools = per_school_summary.sort_values(by = "Overall Passing Rate (%)", ascending = True)
bottom_schools.head()
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_spending = per_school_summary.copy()
```

```python
school_spending["Spending Range"] = pd.cut(per_school_summary["Per Capita Budget ($)"], spending_bins, labels = labels)
school_spending
```

```python
spending_math_scores = school_spending.groupby(["Spending Range"])["Math Average Score"].mean()
spending_reading_scores = school_spending.groupby(["Spending Range"])["Reading Average Score"].mean()
spending_passing_math = school_spending.groupby(["Spending Range"])["Math Passing Rate (%)"].mean()
spending_passing_reading = school_spending.groupby(["Spending Range"])["Reading Passing Rate (%)"].mean()
overall_passing_spending = school_spending.groupby(["Spending Range"])["Overall Passing Rate (%)"].mean()
```

```python
spending_summary = pd.DataFrame({
    "Math Average Score": spending_math_scores,
    "Reading Average Score": spending_reading_scores,
    "Math Passing Rate (%)": spending_passing_math,
    "Reading Passing Rate (%)": spending_passing_reading,
    "Overall Passing Rate (%)": overall_passing_spending
})

spending_summary
```
